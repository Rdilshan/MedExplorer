```python

def threasholding(image):
  img_gray = cv2.cvtColor(image,cv2.COLOR_BGR2GRAY)
  ret,thresh = cv2.threshold(img_gray,120,230,cv2.THRESH_BINARY_INV)
  #plt.imshow(thresh,cmap="gray")
  thresh = cv2.merge((thresh,thresh,thresh))
  return thresh

for dirpath, dirnames, filenames in os.walk(data_dir):
  for filename in filenames:
    img = cv2.imread(os.path.join(dirpath, filename))
    if img is not None:
      threasholding(img)

train_datagen_test = ImageDataGenerator(rescale=1./255)

train_data_test = train_datagen_test.flow_from_directory(data_dir,
                                               batch_size=32,
                                               target_size=(120, 600),
                                               class_mode="categorical",
                                               seed=42)
```

```python
# prompt: used train_data showing random image and label

import matplotlib.pyplot as plt
import random

def show_random_image_and_label(train_data):
  """Shows a random image and its corresponding label from the training data.

  Args:
    train_data: The training data generator.
  """

  batch_images, batch_labels = next(train_data)
  random_index = random.randint(0, len(batch_images) - 1)

  image = batch_images[random_index]
  label = batch_labels[random_index]

  # Assuming categorical labels, find the index of the highest probability
  label_index = label.argmax()
  class_names = list(train_data.class_indices.keys())
  label_name = class_names[label_index]

  plt.imshow(image)
  plt.title(f"Label: {label_name}")
  plt.axis("off")
  plt.show()

show_random_image_and_label(train_data_test)
```

```python

def show_predict_image_and_label(model, img_path, class_names):
  """Shows an image, its predicted label, and its actual label.

  Args:
    model: The trained model.
    img_path: Path to the image.
    class_names: List of class names.
  """

  img = image.load_img(img_path, target_size=(120, 600))
  img = image.img_to_array(img)
  img = np.expand_dims(img, axis=0)
  img = img/255

  prediction = model.predict(img)
  predicted_class = class_names[prediction.argmax()]

  # Extract actual label from image path
  actual_class = img_path.split("/")[-2]

  plt.imshow(img[0])
  if predicted_class == actual_class:
    plt.title(f"Predicted: {predicted_class}, Actual: {actual_class}", color='green')
  else:
    plt.title(f"Predicted: {predicted_class}, Actual: {actual_class}", color='red')
  plt.axis("off")
  plt.show()
```

```python

import os
from tensorflow.keras.preprocessing import image
import numpy as np

def calculate_accuracy_for_folder(model, folder_path, class_names):
  """Calculates the accuracy of the model for a given folder.

  Args:
    model: The trained model.
    folder_path: Path to the folder containing images.
    class_names: List of class names.

  Returns:
    The accuracy of the model for the folder.
  """

  images = glob(f"{folder_path}/*")
  total_images = len(images)
  correct_predictions = 0

  for img_path in images:
    img = image.load_img(img_path, target_size=(120, 600))
    img = image.img_to_array(img)
    img = np.expand_dims(img, axis=0)
    img = img / 255

    prediction = model.predict(img)
    predicted_class = class_names[prediction.argmax()]

    # Extract actual label from image path
    actual_class = img_path.split("/")[-2]

    if predicted_class == actual_class:
      correct_predictions += 1

  accuracy = (correct_predictions / total_images) * 100
  return accuracy


data_dir = "/content/train_data"
for folder_name in os.listdir(data_dir):
  folder_path = os.path.join(data_dir, folder_name)
  if os.path.isdir(folder_path):
    accuracy = calculate_accuracy_for_folder(model_ResNet, folder_path, class_names)
    print(f"Accuracy for {folder_name}: {accuracy:.2f}%")
```
